```python
# import modules
import numpy as np
import matplotlib.pyplot as plt
```

```python
# setting seed
np.random.seed(123)

# initiate walks
walks = []

# start loop to run the simulation a number of times
for x in range(100):    #runs simulation 100 times
#     initiate random_walks
    random_walks = [0]
    for x in range(100):
        step = random_walks[-1]     # initiate step
#         Applying the rules
        dice = np.random.randint(1,7)  # roll dice before taking a step
        if dice <= 2:
            step = max(0, step -1)
        elif dice <= 5:
            step += 1
        else:
            step = step + np.random.randint(1,7)
        if np.random.rand() <= 0.001:
            step = 0
        random_walks.append(step)
    walks.append(random_walks)        
```

```python
# visualizing the simulation
np_trans = np.transpose(np.array(walks))
plt.plot(np_trans)
plt.show()
```

```python
# Select last row from np_trans: ends
ends = np_trans[-1, :]
```

```python
# calculating the probability of reaching steps 60 and above within the number of dice throws
len(ends[ends >= 60]) / 100
```
